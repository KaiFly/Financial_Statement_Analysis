### The Formula:

$$
Fair\;Share\;Price = \frac{Expected\;Future\;Dividends}{(Cost\;of\;Equity - Expected\;Growth)}
$$

<center>Formula</center>

$$
Cost\;of\;Equity = Risk\;Free\;Rate + (Beta \times Market\;Premium)
$$

$$
Market\;Premium = Market\;Return - Risk\;Free\;Rate
$$

$$
Risk\;Free\;Rate = 10\;Years\;Treasury\;Yield
$$

$$
Market\;Return = Return\;of\;the\;S\&P\;500
$$

```python
import yfinance as yf # this is to get the stock data
import pandas as pd # this is for working on data frames
import numpy as np # this is to manipulate the data
```

```python
ticker = yf.Ticker("t") 
# vnindex = yf.Ticker("^VNINDEX.VN")
#vnm = yf.Ticker("VNM.NS") # .NS .HN 
# hist = vnindex.history(period="5d")  # Lấy dữ liệu 1 năm gần nhất
stock = ticker.actions
```

```python
stock_split = stock["Stock Splits"].to_numpy()
stock_split
```

```python
stock_split_replaced = np.where(stock_split == 0, 1, stock_split)
```

```python
stock_split_comp = np.cumprod(stock_split_replaced, axis=0)
```

```python
stock["stocksplit_adj"] = stock_split_comp.tolist()
```

```python
stock["div_adj"] = stock["Dividends"] * stock["stocksplit_adj"]
stock
```

```python
stock['year'] = stock.index.year
stock_grp = stock.groupby(by=["year"]).sum()
stock_grp.tail(20)
```

```python
stock_grp["div_PCT_Change"] = stock_grp["div_adj"].pct_change(fill_method ='ffill')
stock_grp.tail(20)
```

```python
median_growth = stock_grp["div_PCT_Change"].median()
median_growth
```

```python
lst_Div = stock_grp.at[2025,'Dividends']
lst_Div
```

```python
exp_future_div = round(lst_Div * (1 + median_growth),2)
exp_future_div
```

```python
# If you want to use the 10 year treasury yield today use this code
# bond = yf.Ticker("^TNX")
# hist = bond.history(period="today")
# risk_free_rate = round(hist.iloc[0]['Close']/100,4)
# risk_free_rate
```

```python
risk_free_rate = 0.03
# assume the avg return for sp500 = 11%
mkt_return = .11
MKT_Risk_prem = mkt_return - risk_free_rate
MKT_Risk_prem

beta = ticker.info["beta"]
beta

# cost of equity
COE = round(beta * MKT_Risk_prem + risk_free_rate,4)
print(f"COE: {COE}")

#
fair_sharePrice = round(exp_future_div/(COE-median_growth),2)
fair_sharePrice
print(f"fair_sharePrice: {fair_sharePrice}")

#
stock_price = ticker.history(period="1d")
stock_price_close = round(stock_price.iloc[0]['Close'],4)
stock_price_close
print(f"stock_price_close: {stock_price_close}")

# positive means undervalued, negative means overvalued
expected_gain_loss = fair_sharePrice/stock_price_close-1
expected_gain_loss = "{:.0%}".format(expected_gain_loss)
expected_gain_loss
print(f"expected_gain_loss: {expected_gain_loss}")
```
